```python
plt.hist(np.random.rand(100000) , density=True , histtype="step" , bins=100 , color="blue" , label="rand")
plt.axis([-2.5,2.5,0,1.1])
plt.legend(loc="upper left")
plt.title("Random distributions")
plt.xlabel("value")
plt.ylabel("Density")
plt.show()
```

```python
x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
X,Y = np.meshgrid(x,y)
```

```python
Z = X**2 + Y**2
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(X,Y,Z , cmap='viridis')
```

```python
ax.plot_surface(X,Y,Z , cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D plot of Z = X**2 + Y**2')
plt.tight_layout()
plt.show()
```
