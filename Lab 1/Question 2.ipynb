```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
filtered_df = df[['N_total' , 'N_bulk', 'N_surface','R_avg']]
```

```python
print(filtered_df)
```

```python
print(filtered_df.head(20))
```

```python
summary_stats = filtered_df.describe()
print(summary_stats)
```

```python
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
axes[0].hist(filtered_df['N_total'], bins=20, color='blue', alpha=0.7)
axes[0].set_title('N_total Histogram')

axes[1].hist(filtered_df['N_bulk'], bins=20, color='green', alpha=0.7)
axes[1].set_title('N_bulk Histogram')

axes[2].hist(filtered_df['N_surface'], bins=20, color='red', alpha=0.7)
axes[2].set_title('N_surface Histogram')

axes[3].hist(filtered_df['R_avg'], bins=20, color='purple', alpha=0.7)
axes[3].set_title('R_avg Histogram')

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(filtered_df)
plt.show()
```

```python
sns.pairplot(filtered_df, diag_kind='kde')
plt.show()
```

```python
g = sns.pairplot(filtered_df, diag_kind='kde')
g.map_lower(sns.kdeplot, cmap='Blues')
plt.show()
```
